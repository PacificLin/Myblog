```python
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
print ('Classes of each table:')
for table in soup.find_all ('table'):
    print (table.get ('bordor'))
```

```python
tables = soup.find_all('table')
second_table = tables[2] 
second_table
```

```python
# 找到表格行
rows = second_table.find_all('tr')

# 初始化一个空的列表来存储数据
data = []

# 遍历表格行
for row in rows:
    # 找到表格数据单元格
    cells = row.find_all('td')
    
    # 提取每个单元格的文本内容并添加到数据列表
    row_data = [cell.get_text(strip=True) for cell in cells]
    data.append(row_data)
```

```python
# 打印前几行数据
data[len(data) - 1]
```

```python
data[0], data[1]
```

```python
import pandas as pd

# 删除 "基本資料" 和 "相關機構"
data[0] = [col for col in data[0] if col not in ["基本資料", "相關機構"]]
data[0]
```

```python
# 列名列表，包含 "商品代號" 和 "商品名稱"，以及其余的列名
column_names = data[0] + data[1]

# 数据中删除前两行的 "商品代號" 和 "商品名稱" 以及 "基本資料" 和 "相關機構"
data = data[2:]

# 将数据转换为DataFrame
df = pd.DataFrame(data, columns=column_names)
```
